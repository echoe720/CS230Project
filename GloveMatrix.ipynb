```python
def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
embeddings_dict = load_obj("glove_embeddings")
```

```python
datapath = "datasets/C50/"
```

```python
train_df = pd.read_csv(datapath + "train_10_glove.csv")
test_df = pd.read_csv(datapath + "test_10_glove.csv")
```

```python
len(test_df.index)
```

```python
rows = 100
cols = 500
```

```python
def generate_matrices(df): 
    examples = []
    for index, row in df.iterrows():
        matrix = np.zeros((rows, cols))
        author = row["Author"]
        text = row["Text"]
        tokens = nltk.word_tokenize(text)
        i = 0
        for token in tokens:
            word = token
            if word not in embeddings_dict.keys():
                word = 'unk'
            column = embeddings_dict[word]
            matrix[:, i] = column
            i += 1
            if i == cols:
                print("Max length reached.")
                break
        examples.append((author, matrix))
    return examples
```

```python
train_examples = generate_matrices(train_df)
```

```python
train_examples[1][1]
```

```python
test_examples = generate_matrices(test_df)
```

```python
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
save_obj(train_examples, "train_matrices_CNN")
```

```python
save_obj(test_examples, "test_matrices_CNN")
```
