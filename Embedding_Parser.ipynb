```python
import numpy as np
from scipy import spatial
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import pickle
import pandas as pd
```

```python
datapath = "datasets/C50/"
```

```python
train_df = pd.read_csv(datapath + "C50train/C50_train_5_sentences.csv")
```

```python
def make_examples(df):
    examples = {}
    for index, row in df.iterrows():
        author_num = row["Author_num"]
        text = row["Text"]
        if author_num not in examples:
            examples[author_num] = []
        examples[author_num].append(text)
    return examples
```

```python
train_examples = make_examples(train_df)
```

```python
len(train_examples.keys())
```

```python
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
save_obj(train_examples, "examples")
```
