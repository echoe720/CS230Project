```python
EXAMPLE_SIZE = 20
```

```python
train_file_df = pd.DataFrame(columns=["Text", "Author"])
#temp_df = pd.DataFrame()
authors = []
for i, filepath in enumerate(glob.iglob('C50train/*/*.txt')): #prior to running this, one must downlaod the C50.zip file on the main drive

    if i % 500 == 0: 
        print('Analysing', i , ' -> ', ' of Author ', str(filepath.split("/")[1]))
    author = filepath.split("/")[1]
    sentences = []
    with open(filepath) as f:
          text = f.read()
          sentences = sent_tokenize(text) #text.split('\n')
    total_added = 0
    curr_example = "" 
    for j in range(len(sentences)):
        if total_added == EXAMPLE_SIZE:
            total_added = 0
            train_file_df = train_file_df.append({"Text": curr_example, 'Author': author}, ignore_index=True)
            curr_example = ""
        else:
            total_added += 1
            curr_example += sentences[j] + " "

 
print("Done")
```
